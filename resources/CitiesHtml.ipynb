```python
import pandas as pd
from bs4 import BeautifulSoup


file = 'cities.csv'

df = pd.read_csv(file)

sdf = df[['City ID', 'City', 'Country', 'Max Temp', 'Humidity', 'Cloudiness', 'Wind Speed', 'Lat', 'Lng', 'Date']]

cities = sdf.set_index('City ID')
# cities['Humidity'] = pd.Series(["{0:.2f}%".format(val * 100) for val in cities['Humidity']], index = cities.index)
# cities['Cloudiness'] = pd.Series(["{0:.2f}%".format(val * 100) for val in cities['Cloudiness']], index = cities.index)

cities.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import date

# analysis_date = date.today().strftime("%m/%d/%y")
analysis_date = '02/11/2021'

city_data = df.set_index('City ID')
# city_data.head()
city_data.columns = ['City', 'Lat', 'Lng', 'Max Temperature (F)', 'Humidity (%)', 'Cloudiness (%)',
       'Wind Speed (mph)', 'Country', 'Date']

x_values = city_data["Lat"]
for column in city_data[['Max Temperature (F)', 'Humidity (%)', 'Cloudiness (%)', 'Wind Speed (mph)']]:
    y_values = city_data[column]
    plt.scatter(x_values,y_values,edgecolors="black",facecolors="steelblue")
    plt.title(f"City Latitude vs. {column} ({analysis_date})")
    plt.xlabel('Latitude')
    plt.ylabel(column)
    plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
    plt.savefig(f"../assets/Latitude_v_{column}.png")
    plt.show()
```

```python
table = cities.to_html()
html = table.replace('\n', '')
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```
